**3. Show the difference between the average compensation and average salary of professors. Who has the highest difference in their mean?**

```python
data=df1[['Average compensation - full professors','Average salary - full professors','Average salary - associate professors',\
       'Average salary - assistant professors','Average compensation - associate professors','Average compensation - assistant professors']]
s1=data['Average compensation - full professors'].mean()
s2=data['Average compensation - associate professors'].mean()
s3=data['Average compensation - assistant professors'].mean()
a1=data['Average salary - full professors'].mean()
a2=data['Average salary - associate professors'].mean()
a3=data['Average salary - assistant professors'].mean()
prof=['full','associate','assistant']
c=[s1,s2,s3]
s=[a1,a2,a3]
d=pd.DataFrame({'professor':prof,'Compensation':c,'Salary':s})
d.plot(x='professor', y=['Compensation','Salary'], kind="bar")
plt.title('Graph showing the salary and compensation of professors')
plt.show()

m=[s1-a1,s2-a2,s3-a3]
k=0
for i in range(len(m)-1):
  if m[k]>m[i+1]:
    k=k
  else:
    k=i+1
#print(m)
print('The category of professor having highest difference in the mean of salary and compensation are',prof[k],'professors')
```

**7. How are Number of applications received,Number of applicants accepted,Number of new students enrolled related? If a student is chosen at random what is the probaiblity that he enrolled given that he accepted the application?**
