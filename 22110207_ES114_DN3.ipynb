**How does the mean distribution of points of players differ in all sets? Approximately what is the average number of points scored by player1 in all rounds and scored by player 2 in all rounds in the Australian Open Men (2023)? If a set is chosen at random, what is the probability that player 2 scored more (mean as considering the whole set) than player 1 in the Australian Open Men (2023)?**


```python
a=aum[['ST1.1','ST2.1','ST3.1','ST4.1','ST5.1','ST1.2','ST2.2','ST3.2','ST4.2','ST5.2','Player1','Player2']]
x1=a[['ST1.1']].mean()
x2=a[['ST2.1']].mean()
x3=a[['ST3.1']].mean()
x4=a[['ST4.1']].mean()
x5=a[['ST5.1']].mean()

y1=a[['ST1.2']].mean()
y2=a[['ST2.2']].mean()
y3=a[['ST3.2']].mean()
y4=a[['ST4.2']].mean()
y5=a[['ST5.2']].mean()

a=['Player1','Player2']
set1=[x1,y1]
set2=[x2,y2]
set3=[x3,y3]
set4=[x4,y4]
set5=[x5,y5]

#print(set1,set2,set3,set4,set5)

set1=[5.166667,4.984127]
set2=[5.152,4.776]
set3=[4.900826,4.867769]
set4=[4.79661,4.423729]
set5=[4.666667,5.0]

df = pd.DataFrame({'Player': a,'set1':set1,'set2':set2,'set3':set3,'set4':set4,'set5':set5})
df.plot(x='Player', y=['set1','set2','set3','set4','set5'], kind="bar")
plt.grid() # to show grid
plt.title('Graph of mean score scored by players in different sets')
plt.xlabel('Player')
plt.ylabel('Mean score')
plt.legend()
plt.show()

avg1=0
avg2=0
b=[set1,set2,set3,set4,set5]
for i in b:
  avg1=avg1+i[0]
  avg2=avg2+i[1]
avg1=avg1/5
avg2=avg2/5

print('Avg points scored by player 1 in all sets is:',avg1)
print('Avg points scored by player 2 in all sets is:',avg2)

s2w1l=0
c=0
t=0
# calculating number of sets in which player 1 has less mean than player 2
for i in b:
  t=t+1
  if i[0]<i[1]:
    c=c+1
print('Probability of player 2 scoring more in sets is:',c/t)
```

**Who won more matches in each round in the Australian open women(2013)?If a match is chosen at random, what is the probability that player1 has won it? Show the distribution of matches won by players.**

```python
p=auw[['Round','Result']]
d=p.groupby(['Round','Result']).Result.count()

d1 = d.unstack(level=[1])
ax = d1.plot(kind='bar',color=['r','k'])
ax.set_ylabel('Count')
ax.set_xlabel('Round')
ax.set_title('Graph showing the distribution of number\n of matches won by players in different rounds\n')
ax.legend(['Player 2', 'Player 1'])
plt.show()

c=p.groupby('Result').Result.count()
#print(c)
p1=66
p2=61
print('If a match is chosen at random, the probability that player1 wins is:',p2/(p1+p2))

plt.title('\n Pie chart showing the percent of matches won by players\n')
plt.pie([p1,p2],labels=['Player 2','Player 1'],autopct='%1.1f%%',explode=[0.1,0],colors=['m','c'])
plt.show()
```

**Show the distribution of first serve percentage players get in French open women (2013). What is range of precentage is most common? Show the first serve win percentage. What is the maximum first serve win percentage?**

```python
d=fow[['FSP.1','FSW.1','FSP.2','FSW.2']]
plt.figure()
sns.kdeplot(data=d,x='FSP.1',c='r',label='first serve precentage player1')
sns.kdeplot(data=d,x='FSP.2',c='m',label='first serve percentage player2')
plt.xlabel('Percentage of first serve of player 1 and player 2')
plt.ylabel('Density')
plt.title('Graph showing the first serve percentages of player 1 and player 2')
plt.show()

sns.kdeplot(data=d,x='FSW.1',c='c',label='first serve win percentage player1')
sns.kdeplot(data=d,x='FSW.2',c='k',label='first serve win percentage player 2')
plt.xlabel('First serve win percentage of player 1 and player 2')
plt.ylabel('Density')
plt.title('Graph showing the first serve win percentages of player 1 and player 2')
plt.show()

p1m=d[['FSP.1']]
a=p1m.max()
#print(a)
print('Max first serve percent for player 1 is:',82)

p2m=d[['FSP.2']]
b=p2m.max()
#print(b)
print('Max frist serve percent for player 2 is:',91)

p1m=d[['FSW.1']]
a=p1m.max()
#print(a)
print('Max first serve win percent for player 1 is:',54)

p2m=d[['FSW.2']]
b=p2m.max()
#print(b)
print('Max frist serve win percent for player 2 is:',58)
```
